# Continuous Uncertainty Encoding: The Bus Example

---

- **Continuous uncertainty encoding** is a type of data visualizatiion that allows you to represent variability through continuous representation of the data. 

- In each of these encodings, the visual representation of the data provides information about the range of values that measures or estimates mights take and the degree of confidence in that range. 

    - Often used in scientific and engineering contexts where the measurement and estimates can be influenced by a variety of sources of error or uncertainty. 

- By encoding this uncertainty in the data visualization, analysts can better understand the limitations of the data and make more informed decisions based on that info. 



CDF's and PDF's are both specific examples of what we call continuous uncertainty encodings.

These representations of uncertainty are called continuous uncertainty encodings because they involve encoding uncertainty through a contunous distribution rather than a discrete distribution. 



This paper explores methods for presenting uncertainty for a transportation example. 



Suppose you're living in Seattle, Washington. The bus company displays on a digital board the approximate time the bus will arrive at each stop. 

For example, it notes that bus 120 should be at the Downtown Center in about 11 min. So you set your watch to 11 minutes and get a coffee shop. There is a note below the sign. There is uncertainty. 

#### So do I have enough time to get a coffee?



The systems is telling us that the bus will arrive sometime between 0 and 30 min. from now, or even further in the future. The system prediction is that the bus will show up 11 min from now exactly. But there is obviously uncertainty sround that point estimate. 

A more realistic and better representation would be a representation that uses some kind of probability distribution around the point estimate. 

I would acknowledge that the bus might now come exactly 11 minutes from now, but likely between 5 and 20 minutes with the most likely value being about 11 minutes. 

But there is a long tail to the right where perhaps traffic and other events might delay the us. 

The coffee shop is 8 minutes away. Do I have enough time? This display does not provide enough information. 



So we add to our display a mark at 8 minutes and find that 90% of the time the bus comes later than 8 minutes. This helps quantify the risk of getting a coffee.

Am I fine with missing the bus 10% of the time if it means that I can enjoy a coffee 100% of the time? 

This is a probability density plot. It's a helpful way to think about the problem but it comes with its limitations. 

it encodes one estimate of risk, in this case, s 90% likelihood of catching the bus. This may work for some  but not others. Also, the amount of risk we're willing to accept coudl flex depending on the situation (i.e., going to an interview, going to the library, etc.)



Another limitation is that it relies on area to convey the probability. It can be difficult to accuretly read area.

What proportion of the outer circle's area is covered by the inner circle as a value of 0 to 100%? In other words, how much of the big circle does a small spur circle cover?

Most people would estimate 85-90%, while the true proportion is 75%. The reason is that area perception in humans is biased. As things get larger, we don't preceive them as being proportionally as large as what they actually are. 

One option is to use a display that does not reply on area to read probability and it can offer probability for any bus arrival time (i.e., a cumulative distribution function)

For example we wanted to know if our coffee took a little longer to get one day, was it took 10 minutes to purchase and get back to the stop. 



What's the probability that we will miss the bus?

As you can see, the CDF returns a probabilty of 25%. It measn that there is a 25% chance that the bus has already come. 

Sometimes you might have to layer uncertainty upon certainty. 



The CDF at 8 minutes yields 10% probability of missing the bus. 

We already know there is a 25% probability that we missed the bus if our coffee takes 10 minutes. 

So if we go get a coffee, there is approximately a 10-25% probability that I missed the bus. 

CDF is equal to the integral of the PDF. 



If we integrate from negative infinity through a value, in this case 10, we get the cumulative probability for 10 ir less. 

The grammer of graphics can aid in the creation and evaluation of these graphical deistributions.



Starting with the density plot, `time` is mapped to the x channel. The density function $f_bus$ takes `time` as its input and outputs density values which are mapped to the x-position. 

These values are marked through a line. These are explicit encoding. 

The specification we use to create the probability density plot. These are required elements to visually create the PDF rendering of the bus problem. 

One thing you might notice probability isn't in our list of explicit mappings.

That's because probability is implicitly encoded onto area.

The area is created from the explicit encoding of density onto the y channel.

This is one of the limitations of the PDF, that probability is mapped implicitly to area which we have noted is not very easy to read accurately.

So while it does a great job of giving the overall shape of the distribution, it can be difficult to read probability from it.






If we contrast this implementation of the of graphics with the cumulative distribution function, we see similarities and differences.

Once again, we map time to the x channel and a statistical distribution, in this case, the CDF to the y channel, and we mark it as a line. These are similar to the PDF we just saw.

The difference of course is that while probability was implicitly mapped on a PDF, it's explicitly mapped in the CDF. The CDF does not rely on implicit encoding of probability, probability can be read directly from the y-axis.

This is one of the great advantages of the CDF.



The **disadvantage of the CDF** is that its overall shape lacks some of the intuition we get from the pure distribution approach of the PDF.

Density plots generated from a PDF and cumulative distribution function plots are only two examples of continuous uncertainty encodings. We should also note that the visualizations built on a probability density function can take several forms.

We've covered density plots and CDFs, but there are other alternatives.

- The violin plot uses densities to vary the thickness of the display.
- Gradient plots, another type of continuous uncertainty encoding maps density onto opacity.
- CCDFs are again similar to CDFs in that they map probability and not probability density to height.



### Key points:

- In this lecture we explored continuous uncertainty encodings specifically focusing on density plots and CDF plots.
- CDF plots and density plots provide trade-offs for how we visualize uncertainty. Density plots are intuitive, but difficult to read directly.
- We also have other continuous encoding optiions like violin plots, gradient plots, and CCDF plots.
