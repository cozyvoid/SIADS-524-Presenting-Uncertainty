| Choice                                                                                                                | Select? | Explanation                                                                                                                                    |

### 3. Which of the following is not a viable strategy for estimating the standard error?


- Collecting multiple samples from the population.
- Estimate analytically from a single sample
- Estimate it synthetically by generating multiple samples from a single sample.
- Take the standard deviation of the sample and divide by the (sample size x 2).

The answer is:

**Take the standard deviation of the sample and divide by the (sample size x 2).**

Brief explanation:

| Choice                                                                             | Viable? | Explanation                                                                                                                              |
| ---------------------------------------------------------------------------------- | ------: | ---------------------------------------------------------------------------------------------------------------------------------------- |
| **Collecting multiple samples from the population.**                               |     Yes | You could estimate standard error by repeatedly sampling from the population and measuring how much the statistic varies across samples. |
| **Estimate analytically from a single sample**                                     |     Yes | For some statistics, such as the sample mean, standard error can be estimated using a formula, for example: (SE = $\frac{s}{\sqrt{n}}$).   |
| **Estimate it synthetically by generating multiple samples from a single sample.** |     Yes | This describes resampling methods such as the **bootstrap**, where repeated samples are generated from the observed sample.              |
| **Take the standard deviation of the sample and divide by the (sample size x 2).** |      No | This is not a standard or valid standard error formula. For a mean, the usual formula divides by ($\sqrt{n}$), not ($2n$).                   |

### 4. Does standard error of a mean correspond with parametric uncertainty or predictive uncertainty?

The **standard error of a mean** corresponds to **parametric uncertainty**.

It measures uncertainty about an **unknown parameter**, such as the true population mean. For example, if you estimate the mean salary of MADS graduates from a sample, the standard error tells you how much your sample mean would vary across repeated samples.

It is **not predictive uncertainty**, because it is not mainly about uncertainty in a **new individual observation**. Predictive uncertainty would be about something like predicting the salary of one new MADS graduate, which includes both uncertainty in the mean and individual-level variation.

### 5. Does distribution for a new observation correspond with parametric uncertainty or predictive uncertainty?

A **distribution for a new observation** corresponds with **predictive uncertainty**.

It represents uncertainty about what value a **future or unseen individual observation** might take.

Example:

* **Parametric uncertainty:** What is the true mean salary of MADS graduates?
* **Predictive uncertainty:** What salary might a newly observed MADS graduate have?

A predictive distribution usually includes both:

1. uncertainty about the estimated model parameters, and
2. natural variation among individual observations.

### 6. True/False: In bootstrapping, we always resample with replacement.

**True.**

In standard **bootstrapping**, we resample **with replacement** from the original sample to create many “new” simulated samples of the same size.

Briefly: replacement matters because it allows some observations to appear multiple times and others not at all, which mimics repeated sampling variability from the population.

### 9. Which of the following is not true of bootstrapping. [Select all that apply]


- You need a representative sample to create an non-biased bootstrapping distribution.

- The number of replications you choose is should not exceed n=5000.
- Bootstrapping should only be used with sample means.
- Under a frequentist framework, bootstrapping assumes that each datum has the same probability of being selected.
- Under a Bayesian framework, bootstrapping keeps the sample membership constant but let's the probabilities vary.

The statements that are **not true of bootstrapping** are:

* **The resample size should be the size of your original sample or 30, whichever is more.**
* **The number of replications you choose should not exceed n=5000.**
* **Bootstrapping should only be used with sample means.**

| Statement                                                                                                            | Not true? | Brief explanation                                                                                                                                                     |
| -------------------------------------------------------------------------------------------------------------------- | --------: | --------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **You need a representative sample to create a non-biased bootstrapping distribution.**                              |        No | This is generally true. Bootstrapping treats your sample as a stand-in for the population, so a biased/nonrepresentative sample can produce biased bootstrap results. |
| **The resample size should be the size of your original sample or 30, whichever is more.**                           |       Yes | Standard bootstrap resamples are usually the **same size as the original sample**, not “original sample or 30, whichever is more.”                                    |

More replications can be used, though there are diminishing returns.                          |

### 10. True/False. As sample size increases, the difference between a Bayesian bootstrap and a classic (frequentist) bootstrap minimizes.

**True.**

As the sample size increases, the **Bayesian bootstrap** and the classic **frequentist bootstrap** tend to give more similar uncertainty estimates.

Brief explanation: both methods use the observed sample to approximate uncertainty. The frequentist bootstrap resamples observations with replacement, while the Bayesian bootstrap keeps the observed data fixed but varies the weights/probabilities assigned to observations. With larger samples, the practical difference between these approaches usually becomes smaller.
