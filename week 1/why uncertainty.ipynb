### What do we mean by "uncertainty"?

**uncertainty, n.**

*The quality of being uncertain in respect of duration, continuance, occurrence, etc.; liability to chance or accident. Also, the quality of being indeterminate as to magnitude or value; the amount of variation in a numerical result that is consistent with observation.*

Oxford English Dictionary

Imagine the MADS program seeks to understand the incoming programming experience of new students. -> $\mu$

The program asks **3** new students how many years experience of Python programming experience they have. 

**The sample consisted of {0, 2, 3}.** -> $\bar{x}$

**The sample mean is 1.7 years.**

> How certain are we that our sample mean is the population mean?



### Uncertainty is Everywhere

*"Nearly every dataset we work with has some uncertainty, and whether and how  we choose to represent this uncertainty can make a major difference in how accurately our audience perceives the meaning of the data."*

- Claus O. Wilke, *Fundamentals of Data Visualization*

### The Uncertainty to Uncertainty

| Approach | Example |
|---|---|
|Ignore|I will be home at 11:00."|
|Imply it|"I will be home around 11:00."|
|Express it|"I will be home between 10:45 and 11:15."|

### Course Goals 

1. Learn the statistical fundamentals of uncertainty.
2. Learn the visual communicaiton fundamentals for presenting uncertainty.
3. Practice the writing fundamentals for describing uncertainty in statistical results. 

### Key Points:

- Anytime we cannot be certain about something, uncertainty  becomes an issue. 
- Statistics is largely about creating estimates under uncertainty. 
- Uncertainty is often underrepresented in statistical training and in business analytics/reporting. 
- In this course, you will learn about the statistics, visualization, and communication of uncertainty.
