### 4. The t-distribution was designed specifically for when:

- **Sample sizes are small.**
- The population variance is known.
- The z-score does not converge over the sample.
- when we need a distribution that has higher peaks and thinner tails than the normal distribution. 

---

The t-distribution was created by William Sealy Gosset. He was a chemist and statistician who worked for the Guinness Brewing company and is best know for his work in statistical techniques for small samples.

The **t-distribution** was designed for situations where the **sample size is small** and the population standard deviation/variance is unknown, so we estimate uncertainty using the sample standard deviation.

It also has **heavier tails than the normal distribution**, not thinner tails, which accounts for the extra uncertainty from small samples.

### 5. As the degrees of freedom (DOF) increase, the t-distribution

- **Looks more like a normal distribution (converges).**
- Looks less like the normal distribution (diverges).
- Tails increase in thickness.
- Remains unchanged.

---

As DOF increase, the t-distribution converges to the normal distribution.

As the degrees of freedom increase, the t-distribution approaches the normal distribution.

The reason is that with more degrees of freedom, there is less extra uncertainty from estimating the population standard deviation using the sample standard deviation. The tails become thinner and closer to the normal curve.

- A and B have the same number of parameters.

---

The correct answer is that they have the same number of parameters (1) because we are examining the t-distribution. The two distributions must be generated from samples with different sizes, which generates the difference in DOF (n-p).

### 9. Which of the following is an accurate interpretation of a 95% confidence interval?

- There is a 95% probability that the population parameter is within the bounds of the confidence interval.
- There is a 5% chance that the population parameter is within the bounds of the confidence interval.
- **The procedure use to create this confidence interval will capture the true population parameter 95% of the time.**
- The procedure use to create this confidence interval will capture the true population parameter 5% of the time.
- None of the above

---


The level of confidence (95%) is not confidence in any one interval, but in the procedure that creates the intervals. 

A 95% confidence interval does not mean there is a 95% probability that the true parameter is inside this specific interval.

Instead, it means that if we repeatedly took samples and built confidence intervals the same way, about 95% of those intervals would contain the true population parameter.
