### Suppose we have a 95% confidence interval that we want to interpret.

Here, we have two interpretations of the 95% confidence interval. 

1. There is a 95% chance that the true population parameter lies within the range. 

2. The procedure used to create this confidence interval will capture the true population parameter 95% of the time. Suppose we have 100 samples from a population and compute a 95% confidence interval for each sample, about 95 of the 100 confidence intervals generated should contain the true population parameter. 



Which do you think is correct?

... 2. is correct, though it is not nearly as clear and intuitive as the first. if I poll most users and many data scientists about the true meaning of a confidence interval, you'd most likelt get a definition closer to option 1. 

#### Why is the second definiton correct?

It has a lot to do with the frequentist view of statistics and some philosophical matters.

However, the basics is that under the frequentist assumptions, the confidence interval is about the process of calculating the confidence interval and not about any individual interval. So we cannot say that there's a 95% chance that our specific confidence interval contains the true population parameter. 

From a frequentist perspective, the interval either does or does not contain the parameter. 

Here's an example to help illustrate the process and perhaps identify a path forward for communicating and interpreting confidence intervals. 

In the serine theory website, you can explore confidence intervals. It's a helpful visual confirmation for how we interpret CI's. 

Suppose we start with a sample that we want to use to estimate the population mean. That's the population parameter we're interested in this case. 



If we want to examine a 90% confidence interval for a mean, we can repeatedly sample from the population and construct a 90% confidence interval around each sample. 



You can see that the site also provides a running total of the confidence intervals containing the population mean and those that do not. 143 out of a possible 1431 do not contain the population parameter, which is what we'd expect from a 90% confidence interval. 10% of the confidence intervals do not capture the population parameter. 

Some may think, if 90% of the confidence intervals generated capture the population mean and I randomly select one, can't I say I'm 90% sure that the CI will capture the population mean? Unfortunately, the answer is no. 

Once we move from an infinite number of confidence intervals generated by the process to a single confidence interval. We can't say we're 90% sure that that particular confidence interval contains the population mean because "one does not equal infinity".



The confidence level is about an infinite process, not about an individual confidence interval. 

What can we say then about these confidence intervals?

Over the years, I've developed a bit of a workaround to this philosophical conundrum. 

When I build confidence intervals, often 90%, I liket o say it represents a reasonable range for our population parameter. I don't specifically mention 95% because that specificity gets us into trouble as illustrated. 

Knowing that 95% of confidence intervals generated under this process should capture the population parameter. The specific CI generated is a reasonable range for the population parameter. Even though we don't know if the actual CI we have does contain the population parameter. 

### General Thoughts on Confidence Intervals 

1. The greater the level of confidence you desire, the wider the confidence interval will be. By this point, you should understand the intuition and it's fairly clear. The more certain we want to be to capture the population parameter with our confidence interval, the wider we must go. 



2. As our uncertainty that is our standard error increases, so does the width of our confidence intervals. On-screen, we have three confidence intervals that are all 95% CI's. The confidence intervals get wider as the standard errors increase. That's because we're multiplying our critical values by larger numbers when the standard errors increase. 



3. One way we measure the quality of an uncertainty visualization is how easy it is to ignore the uncertainty. Uncertainty researchers find that it's a lot easuer for people to ignore uncertainty if we just have a point estimate and a single confidence interval. When we include a second confidence interval, for example, layering a 66th and 95th confidence interval together, it's more likely that the end user will appreciate the underlying uncertainty around the point estimate. 



4. One common mistake in calculating confidence intervals is using the wrong quantile values. A 95% confidence interval has 2.5% of the area of the curve at each tail. In the picture above, we have a vertical reference line at the 2.5% and 97.5% lines. Between the 2.5% and 97.5% percentile is 95% of the distribution mass, which is what constitures our 95% confidenence interval. 



Why is outside that mass is 5% of the data divided between the upper and lower tail? 

A mistake that some students make is that they may put 5% of the mass in each tail for a 10% total outside of the confidence interval. This would correspond to a 90% confidence interval as opposed to a 95% confidence interval.
